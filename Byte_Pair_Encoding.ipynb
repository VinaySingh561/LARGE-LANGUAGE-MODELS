```python
import importlib
import tiktoken

print("tiktoken version : ", importlib.metadata.version("tiktoken"))
```

```python
tiktoken.__version__
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text =(
    "Hello, my name is Vinay what is your name? <|endoftext|> In the sunlit terraces of some unkown"
    "heyyyy are you doing great?"
)

ids = tokenizer.encode(text,allowed_special={"<|endoftext|>"})
print(ids)
```

```python
tokenizer.decode(ids)
```
