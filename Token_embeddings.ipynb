```python
# !pip install gensim
import gensim.downloader as api

model = api.load("word2vec-google-news-300")
```

### example of a word as a vector

```python
word_vector = model
print(word_vector['Laptop'])
```

```python
print(word_vector.most_similar(positive=['king','woman'],negative = ['man'],topn=10))
```

```python
## check the similarity betweew two words
print(word_vector.similarity("bat","ball"))
print(word_vector.similarity("men", "table"))
```

### Creating Token Embeddings

```python
import torch
input_ids = torch.tensor([2,5,1,3])
```

```python
vocab_size = 6
dim  = 3
torch.manual_seed(123)
embedding_layer = torch.nn.Embedding(vocab_size,dim)
```

```python
embedding_layer.weight
```

```python
## if you want to know vector for input_ids
print(embedding_layer(input_ids))
```
